```python
# to print images
def show_image(image):
    plt.figure(figsize=(10, 10))
    plt.imshow(image, cmap="gray")
    plt.show()
```

```python
n = 7260
video = matrix_video[0:n,:,:,:]
video_out = np.ones_like(video)[:,:,:,0]
background_sub = cv2.createBackgroundSubtractorMOG2()
i = 0
video = cv2.VideoCapture("../videos/short.mp4")
out = cv2.VideoWriter('../videos/output_rest1.mp4',cv2.VideoWriter_fourcc(*'mp4v'), 30, (1620,960))
while i < n:    
    # read the video frame
    succes, frame = np.array(video.read())
    # if not 
    # show_image(frame)
    # if there are no more frames to show, break the loop
    
    # rearange the color channels
    # frame = frame[:, :, ::-1]
    # apply the background subtractor to the frame
    mask = background_sub.apply(frame)
    _, mask = cv2.threshold(mask, 250, 255, cv2.THRESH_BINARY)
    kernel = np.ones((2,2), np.uint8)
    mask = cv2.erode(mask, kernel, iterations = 1)
    mask = cv2.dilate(mask, kernel, iterations = 2)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    frameCopy = frame.copy()
    for cnt in contours:
        if cv2.contourArea(cnt) > 650:
            x, y, width, height = cv2.boundingRect(cnt)
            cv2.rectangle(frameCopy, (x , y), (x + width, y + height),(0, 0, 255), 2)
            cv2.putText(frameCopy, 'Car Detected', (x, y-10), cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 0.3, (0,255,0), 1, cv2.LINE_AA)
    mask = np.transpose(np.array((mask,mask,mask)),(1,2,0))
    print(frameCopy.shape)
    stacked_frames = np.hstack((frame, mask, frameCopy))
    print(stacked_frames.shape)
    out.write(stacked_frames)
    # apply the opening morphological operation to the mask to remove the noise
    
    # opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations=1)

    # closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel, iterations=1)
    # print(opening)
    # get the contours of the moving objects
    # contours, _ = cv2.findContours(opening, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)
    # stack the frame horizontally
    # hstacked_frames = np.hstack((mask, opening))
    # cv2.imshow("1",hstacked_frames)
    i += 1
    # resta = cv2.cvtColor(frame,cv2.COLOR_RGB2GRAY)-cv2.cvtColor(fons,cv2.COLOR_RGB2GRAY)
    # show_image(resta)
    # show_image(hstacked_frames)
    # print(resta.shape)
    # out.write(resta)
    
    # video_out[i,:,:] = hstacked_frames
    # if the 'q' key is pressed, stop the loop
    if cv2.waitKey(30) == ord("q"):
        break
video.release()
out.release()
```
